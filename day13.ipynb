# IQR Method

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR = Q3-Q1
LB=Q1-1.5*IQR
UB=Q3+1.*IQR
outliers=data[(data<(LB)) | (data>(UB))]
```

```python
outliers
```

# Z-Score 

```python
sample=pd.DataFrame({
    "ages":[10,11,12,13,14,15,50]
})
```

```python
sample
```

```python
stats.zscore(sample["ages"])
```

```python
z_scores=np.abs(stats.zscore(sample["ages"]))
```

```python
z_scores
```

```python
outliers=sample[z_scores>2]
outliers
```

```python
plt.scatter(range(len(sample)),sample["ages"])
plt.xlabel("Index")
plt.ylabel("Ages")
plt.title("Scatter plot showing Outlier")
plt.show()
```

```python
plt.hist(sample["ages"],10)
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.title("Histogram showing Outlier Effect")
plt.show()
```
