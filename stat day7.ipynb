# Concate Data Frames

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city":["mumbai","banglore","hyderabad"],
    "Temperature":[32,45,35]
})
india_weather
```

```python
import pandas as pd
US_weather = pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "Temperature":[32,45,35]
})
US_weather
```

```python
df=pd.concat([india_weather,US_weather])
df
```

```python
df=pd.concat([india_weather,US_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,US_weather],axis=1)
df
```

# Merge DataFrames

```python
temp_df=pd.DataFrame({
    "city":["Hyderabad","Vijayawada","Visakhapatnam","Chennai"],
    "temperature":[32,45,30,35]})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "city":["Hyderabad","Vijayawada","Visakhapatnam","Banglore"],
    "humidity":[42,75,50,90]})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on="city")
df
```

```python
df=pd.merge(temp_df,humidity_df,on="city")
df
```
